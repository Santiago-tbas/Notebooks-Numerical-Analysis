Ejercicio #1 (Epsilon de máquina)

 + Realizar un programa que calcule el epsilon de máquina.




```python
a=1 #valor inicial#
p=1 #contador de pasos#
e=1 #epsilon a reducir#
condicion=True #condición que va a verificar while #

while condicion:
  p=p+1 #contador de pasos#
  a=a+e #definición de nuestra condición a verificar#
  condicion=a<a+e #verificamos que si el nuevo valor de a  (con el epsilon de máquina) es distinto (o no) al inicial#
  e=e/2 #cambiamos el valor de epsilon hasta que se verifique que la condición sea verdadera#

print("Epsilon de máquina:",e)
print("Números de pasos necesitados para calcular el epsilon de máquina:",p)
```

Ejercicio #2

 + Realizar un programa que calcule pi con una precisión de $10^{-8}$ usando un ciclo ***for***.

  $\sum_{k=0}^{\infty}\frac{(-1)^k}{2k+1}$

 + Calcular el error relativo usando la constante de Numpi para Pi

 + Graficar la secuencia del error de truncamiento $R_n$ y verifique graficamente que $R_n = O(\frac{1}{n})$

```python
from numpy.lib.function_base import append

#importamos algunos paquetes#
import numpy as np
from matplotlib import pyplot as plt

#definimos nuestras condiciones y variables#

precision=10**(-5) #Bajamos la precisión pues la interfaz de Colab no procesa una presición inicial de 10^{-8}#
S=4 #valor incial de aproximación que vamos a reducir usando la precisión#
n=int(2*(1/precision)) #entero hasta donde se va a calcular la serie (valor de troncamiento)#
error_troncamiento=abs(S-np.pi)

x=[0]
y=[error_troncamiento]
z=[]

for k in range(n):

#Mejoramos nuestra aproximación reduciendo el error en cada caso#
  S=S+((-1)**(k+1))*4/(2*(k+1)+1)
  error_troncamiento=abs(S-np.pi)

#Registramos los datos que nos serviran para el grafico#
  x.append(k+1)
  y.append(error_troncamiento)
  z.append(1/(k+1))

print("Valor de Pi: ",S)
print("Error de troncamiento: ",error_troncamiento)
print("Precisión obtenida: ", precision)

#Error relativo#
error_relativo=error_troncamiento/abs(np.pi)
print("Error relativo: ", error_relativo)

#Gráficos#
plt.plot(x,y)
x.remove(0)
plt.plot(x,z)
plt.axis([0,25,0,1])

```

Podemos observar que $R_n = O(\frac{1}{n})$
